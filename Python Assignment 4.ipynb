## Creating a Class (Initializing/Declaring)

## The \__init\__() Method <br>
<p>This method is used in almost every created class, and called only once upon the creation of the class instance. This method will initialize all variables needed for the object.</p>

```python
# dunder = double underscore
class Car():

  def __init__(self,wheels,color):
     self.wheels = wheels
     self.color = color

ford = Car(4,'red')
chevy = Car(6,"black")

print(ford.wheels)
print(chevy.color)
```

## Class Attributes <br>
<p>While variables are inside of a class, they are referred to as attributes and not variables. When someone says 'attribute' you know they're speaking about a class. Attributes can be initialized through the init method, or outside of it.</p>

##### Initializing Attributes

```python
class Toy():
  kind = 'car'

  def __init__(self,rooftop,horn,wheels = 4):
    self.rooftop = rooftop
    self.horn = horn
    self.wheels = wheels

tonka = Toy(1,1)
hotwheels = Toy("convertible",1,3)


class Square():
  def __init__(self,length):
    self.length = length
    self.area = length ** 2
    self.perimeter = length * 4
    self.urmom = print("ur mom's a square")

  def urMom(self):
    print("ur mom is a lovely woman")

square1 = Square(3)
square1.urMom()
```

## Class Methods <br>
<p>While inside of a class, functions are referred to as 'methods'. If you hear someone mention methods, they're speaking about classes. Methods are essentially functions, but only callable on the instances of a class.</p>

##### Creating

```python
class ShoppingBag():

  '''
    the ShoppingBag class will have handles, capacity,
    and items to place inside.

    Attribute for the class:
    - handles: expected to be an integer
    - capacity: expected to be an integer
    - items: expected to be a list
  '''

  def __init__(self,handles,items = [], capacity = 10):
    self.handles = handles
    self.items = items
    self.capacity = capacity

  def showShoppingBag(self):
    if len(self.items) == 0:
      print("Your bag is empty!")
    else:
      print("You have these items in your bag:")
      for item in self.items:
        print("- " + item.title())

  def showCapacity(self):
    remainingCapacity = (self.capacity - len(self.items))
    print(f"Your remaining capacity is: {remainingCapacity}.")

  def addToShoppingBag(self):
    if len(self.items) == self.capacity:
      print("Your cart is full!")
    else:
      product = input("What would you like to add? ")
      self.items.append(product)

def showHandles(self):
  print(f"You have {self.handles} handles on your bag")

wholefoodsbag = ShoppingBag(2)
```

```python
# Create a class called cart that retains items and has methods to add, remove, and show

class Cart():
  '''
    the Cart class will have items to place inside
    and methods to add, remove and show cart.

    Attribute for the class:
    - items: expected to be a list
  '''

  def __init__(self,items=[]):
    self.items = items

  def showCart(self):
    if len(self.items) == 0:
      print("Your cart is empty!")
    else:
      print("\nYou have these items in your cart:")
      for item in self.items:
        print("- " + item.title())

  def addToCart(self):
      addItem = True
      while addItem == True:
        product = input("What would you like to add? ")
        self.items.append(product)
        addMore = input(f"Adding {product} to your cart. Do you want to add more? \n y | n ")
        if addMore.lower() == "y":
          continue
        elif addMore.lower() == "n":
          addItem = False
        else:
          ("Please enter a valid response.\n y | n ")

  def removeFromCart(self):
      print("What would you like to remove from your cart?")
      for item in self.items:
        print("- " + item.title())
      product = input(" ")
      self.items.remove(product)

jenn = Cart()
def run():
  while True:
    response = input("\nWhat would you like to do? \n Add | Show | Remove | Quit ")

    if response.lower() == 'quit':
      jenn.showCart()
      print("Thanks for shopping!")
      break
    elif response.lower() == 'add':
        jenn.addToCart()
    elif response.lower() == 'show':
        jenn.showCart()
    elif response.lower() == 'remove':
        jenn.removeFromCart()
    else:
      print("Please enter a valid response.\n Add | Show | Remove | Quit ")

run()
```
